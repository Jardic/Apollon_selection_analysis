# Apollon reselection full-length sequence logo
In this notebook I create the sequence logo of full length apollon from the randomly mutagenized library (21%). As we dont have the initial library sequenced, initial frequencies used are theoretical (0.79 for an unchanged base, 0.07 for a changed base). No normalization was applied.

```python
import pandas as pd
from ivstools import tools
import matplotlib.pyplot as plt
```

```python
df_counts_1nt = pd.read_csv('/home/kurfurst/Documents/projects/mv/apollon/apollon_reselection/analysis/pos_base_cfe/apollon_reselection_1nt_counts.tsv', sep='\t', index_col=0)
```

```python
df_info_1nt = tools.getInformationMatrix(df_counts_1nt)
```

```python
df_info_1nt['conservation'] = df_info_1nt.sum(axis=1)
```

```python
tools.getLogo(df_info_1nt, [x for x in range(0, 85)], figure_file='apollon_reselection_fl_logo_PATENT.svg', suppress_const=False, colorscheme='grays')
```
